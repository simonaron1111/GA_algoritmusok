````python
def bit_not(x):
  return ''.join(['1' if letter == '0' else '0' for letter in x])

def deceptiv_fuggveny(fitnesses, x):
  y = bit_not(x)
  return fitnesses[y]

def hatvany_halmaz(symbols, length):
  def _hatvany_halmaz(k, symbols, length, sol):
    if k == length:
      return [sol]
    result = []
    for symbol in symbols:
      result.extend(_hatvany_halmaz(k + 1, symbols, length, sol + symbol))
    return result
  return _hatvany_halmaz(0, symbols, length, "")

def flip_zeros_to_ones(schema):
  return ''.join(['1' if letter == '0' else letter for letter in schema])

def avg_fitness_of_schema(fitnesses, symbols, schema):
  fitting_strings = []
  for string in hatvany_halmaz(symbols, len(schema)):
    matches = True
    for i in range(len(schema)):
      if schema[i] != '*' and schema[i] != string[i]:
        matches = False
        break
    if matches:
      fitting_strings.append(string)
  if len(fitting_strings) == 0:
    return 0
  return sum([deceptiv_fuggveny(fitnesses, string) for string in fitting_strings]) / len(fitting_strings)

def check_fitness(fitnesses, length = 4):

  all_keys = hatvany_halmaz(['0', '1'], length)

  for block in all_keys:
    print(block, deceptiv_fuggveny(fitnesses, block))

  maximum_fitness = max(deceptiv_fuggveny(fitnesses, x) for x in all_keys)
  best_points = [block for block in all_keys if deceptiv_fuggveny(fitnesses, block) == maximum_fitness]
  print("Globális maximum érték:", maximum_fitness)
  print("Globális maximum helye(i):", best_points)

  symbols = set(['0', '1', '*'])
  violations = []

  for schema in hatvany_halmaz(symbols, length):
    if schema == "*" * length:
      continue
    if '0' not in schema:
      continue # nothing to swap
    if '*' not in schema:
      continue # already specified all bits

    schema_flipped = flip_zeros_to_ones(schema)
    avg_fitness_original = avg_fitness_of_schema(fitnesses, symbols, schema)
    avg_fitness_flipped = avg_fitness_of_schema(fitnesses, symbols, schema_flipped)

    if avg_fitness_flipped <= avg_fitness_original:
      violations.append((schema, schema_flipped, avg_fitness_original, avg_fitness_flipped))

  if len(violations) == 0:
    print("Minden vizsgált sémára teljesül:")
    print("  a 0→1 csere növeli a szkéma átlagos fitnesszét")
  else:
    print("Találtunk ellenpéldá(kat):")
    for schema, schema_flipped, avg_fitness_original, avg_fitness_flipped in violations:
      print(schema, "->", schema_flipped, ": átlag", avg_fitness_original, "->", avg_fitness_flipped)
````
